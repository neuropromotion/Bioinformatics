```python
from Bio import Entrez, SeqIO
def download_gb(name):
  Entrez.email = "neuro.promotion@gmail.com"
  handle = Entrez.efetch(db="protein", id=name, rettype="gb", retmode="text")
  record = SeqIO.read(handle, "genbank")
  handle.close()
  SeqIO.write(record, "data.gb", "genbank")
```

```python
import re
def extract_features(observation):
  Entrez.email = 'A.N.Other@example.com'
  start = int(((observation[0][4:]).split(':'))[1])
  end = int(((observation[0][4:]).split(':'))[2])
  orf_number = observation[0][4:].split(':')[0]
  if start < end:
    strand = 1
  else:
    strand = -1
    start, end = end, start
  handle = Entrez.efetch(db="protein", id=observation[1], rettype="gb")
  record = SeqIO.read(handle, "genbank")

  perc_ident = observation[2]
  e_value = observation[3]
  positives = observation[4]

  download_gb(observation[1])
  gene = None
  organism = None
  product = None
  # Ничего лучше не придумал как скачать генбанк файл и из него уже достать информацию
  with open('data.gb', 'r') as ptr:
    for line in ptr:
      if '/gene=' in line:
        gene = line
      if 'ORGANISM' in line:
        organism = line[12:].strip()
      if '/product=' in line:
        product = line

  if gene:
    gene = re.findall(r'"([^"]*)"', gene)[0] # регулярное выражение для того что бы достать название гена из кавычек
  else:
    gene = 'Unknown'
  product = re.findall(r'"([^"]*)"', product)[0] # регулярное выражение для того что бы достать название белка из кавычек

  sequence = record.seq
  note = 'Organism: ' + organism + '; ' + 'Identities: ' + str(perc_ident) + '; ' + 'E-value: ' + str(e_value) + '; ' + 'Positives: ' + str(positives) + '.'

  result =  {'orf':orf_number,
            'gene':gene,
            'product':product,
            'translation':sequence,
            'start':start,
            'end':end,
            'strand':strand,
            'note':note}
  handle.close()
  return result
```

```python
def extract_nucleotides(observation):
  Entrez.email = 'A.N.Other@example.com'
  start = int(((observation[0][11:]).split('-'))[0])
  end = int(((observation[0][11:]).split('-'))[1])
  if start < end:
    strand = 1
  else:
    strand = -1
    start, end = end, start

  handle = Entrez.efetch(db="nucleotide", id=observation[1], rettype="gb")
  record = SeqIO.read(handle, "genbank")

  perc_ident = observation[2]
  e_value = observation[3]
  id = observation[1]
  product = str(record.description)
  note = str(record.annotations['organism']) + '; identity: ' + str(perc_ident) + '; e-value: ' + str(e_value)


  result =  {'genomic_region_name':id,
            'product':product,
            'start':start,
            'end':end,
            'strand':strand,
            'note':note}

  handle.close()
  return result
```

```python
record = SeqRecord(seq,
                   id='hse666',
                   name='fragment15',
                   description="part of DNA",
                   annotations={"molecule_type": "DNA", "organism":"Corynebacterium diphtheriae"})

for sample in proteins:
  feature = SeqFeature(FeatureLocation(start=sample['start'], end=sample['end'], strand=sample['strand']),
                      type='CDS',
                      qualifiers={'gene':sample['gene'],
                                  'product':sample['product'],
                                  'translation':sample['translation'],
                                  'note':sample['note']})
  record.features.append(feature)

for sample in tRNA:
  feature = SeqFeature(FeatureLocation(sample['start'], sample['end'], strand = sample['strand']),
                      type='tRNA',
                      qualifiers={'rna':sample['rna'],
                                  'Sequence' : sample['seq'],
                                  'note': sample['note']})
  record.features.append(feature)

for sample in nucleotides:
  feature = SeqFeature(FeatureLocation(sample['start'], sample['end'], strand=sample['strand']),
                      type='non translated',
                      qualifiers={'genomic_region_name':sample['genomic_region_name'],
                                  'product':sample['product'],
                                  'note':sample['note']})
  record.features.append(feature)


# Save as GenBank file
output_file = open('fragment15.gb', 'w')
SeqIO.write(record, output_file, 'genbank')
```
